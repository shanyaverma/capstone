```python
X.append(a)
X=np.array(X)
```

```python
Y=[]
Y.append(1)
Y=np.array(Y)
```

```python
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot 
from matplotlib.pyplot import imshow
#from IPython.display import Image 
from IPython import display
from PIL import Image
import numpy as np

# configure batch size and retrieve one batch of images
for i in range(0, 2):
    pyplot.subplot(330 + 1 + i)
    datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
# fit parameters from data
    datagen.fit(X)
    for X_batch, Y_batch in datagen.flow(X, Y, batch_size=1):
        # create a grid of 3x3 images
        
        for j in range(0, 1):
            %matplotlib inline
            img=Image.fromarray(X_batch[j].astype('uint8'),'RGB')
            imshow(np.asarray(img))
            img.save("somepic.jpg") 

            #Image(X_batch[j].reshape(a.shape[0],a.shape[1],a.shape[2])))
            #pyplot.imshow(X_batch[j].reshape(a.shape[0],a.shape[1],a.shape[2]))
            # show the plot
            #pyplot.show()
        break
        
```
